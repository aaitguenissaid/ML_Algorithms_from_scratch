```python
from gradient_descent_2d import Gradient_descent, PlotFunc2D
```

```python
model1 = Gradient_descent(-5, 0.1, 20, "p")
model1.fit()
PlotFunc2D(-5, 5, "Quadratic", "p", model1.wdata, model1.Ldata)

model2 = Gradient_descent(2, 0.2, 20, 'c')
model2.fit()
ax = PlotFunc2D(-5, +5, "Cubic", 'c', model2.wdata, model2.Ldata)

model3 = Gradient_descent(4, 0.02, 50, 'l')
model3.fit()
PlotFunc2D(-5, +5, "Exponential", 'l', model3.wdata, model3.Ldata)

model4 = Gradient_descent(0.5, 0.2, 25, 'e')
model4.fit()
PlotFunc2D(-5, +5, "cosine", 'e', model4.wdata, model4.Ldata)

model5 = Gradient_descent(4, 0.5, 18, 'b')
model5.fit()
PlotFunc2D(0.1, 5, "logarithmic", 'b', model5.wdata, model5.Ldata)

model7 = Gradient_descent(2.3, 0.01, 20, 'r')
model7.fit()
ax = PlotFunc2D(-5, 2.5, "Polynomial", 'r', model7.wdata, model7.Ldata)
```
